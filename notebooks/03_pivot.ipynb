# 03 - Pivot (Estructura)

```python
import pandas as pd
from pathlib import Path
```

```python

base_path = Path(r'C:/Users/rroman/OneDrive - Empresa Eléctrica de Guatemala, S.A. (EEGSA)/Documents/Unidad/2026/Capacitación/Data Science y Machine Learning/proyecto final/data')
input_path = base_path / 'processed' / 'dataset_medidores.parquet'
output_path = base_path / 'processed' / 'dataset_pivot.parquet'

df = pd.read_parquet(input_path)

df_pivot = df.pivot_table(
    index=["Meter ID", "Datetime"],
    columns="Units",
    values="Value",
    aggfunc="first"
).reset_index()

df_pivot.columns.name = None

df_pivot.to_parquet(output_path, index=False)

print(df_pivot.shape)
df_pivot.head()
```
